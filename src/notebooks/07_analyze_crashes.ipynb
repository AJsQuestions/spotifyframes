# ðŸ” Spotify Crash Analysis

Analyze playback errors and crash patterns from exported data. Demonstrative only; sync via CLI or dashboard.

**Prerequisites:** Sync first. Optional: place Spotify Technical Log Information in `data/`.

**Views:** Error counts, error codes, fatal vs non-fatal.

## Setup & load

```python
from pathlib import Path
from notebook_helpers import setup_project, get_data_dir, analyze_crashes, analyze_error_patterns

PROJECT_ROOT = setup_project(Path("../..").resolve())
DATA_DIR = get_data_dir(PROJECT_ROOT)

crash_data = analyze_crashes(DATA_DIR)
errors_df = crash_data.get('playback_errors')
```

## Error patterns

```python
if errors_df is not None and len(errors_df) > 0:
    patterns = analyze_error_patterns(errors_df)
    print(f"Total errors: {patterns.get('total_errors', 0):,}")
    print(f"Fatal errors: {patterns.get('fatal_errors', 0):,}")
    print("Error codes:", patterns.get('error_codes', {}))
else:
    print("No playback error data. Sync and/or place Technical Log Information in data/.")
```

## Error code distribution

```python
if errors_df is not None and len(errors_df) > 0 and 'message_error_code' in errors_df.columns:
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots(figsize=(6, 3))
    errors_df['message_error_code'].value_counts().plot(kind='bar', ax=ax)
    ax.set_xlabel('Error code')
    ax.set_ylabel('Count')
    ax.set_title('Playback errors by code')
    plt.tight_layout()
    plt.show()
```
