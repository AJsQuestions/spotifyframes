## Setup & load

```python
from pathlib import Path
from notebook_helpers import setup_project, get_data_dir, identify_redundant_playlists
from notebook_helpers import build_consolidation_suggestions, build_consolidation_strategies

PROJECT_ROOT = setup_project(Path("../..").resolve())
DATA_DIR = get_data_dir(PROJECT_ROOT)

redundancy = identify_redundant_playlists(DATA_DIR, exclude_auto_generated=True)
consolidation = build_consolidation_suggestions(redundancy)
strategies = build_consolidation_strategies(redundancy, consolidation)
```

## Consolidation suggestions

```python
suggestions = consolidation.get('consolidation_suggestions', [])
print(f"Found {len(suggestions)} suggestion(s).")
for s in suggestions[:15]:
    print(f"  {s.get('action', '')}: {s.get('playlist_name', '')} ‚Äî {s.get('reason', '')}")
```

## Similar-playlist strategies

```python
similar = strategies.get('similar_consolidation_candidates', [])
print(f"Similar-playlist candidates: {len(similar)}")
for c in similar[:10]:
    print(f"  {c.get('playlist1', '')} ‚Üî {c.get('playlist2', '')} ({c.get('similarity', 0):.0f}%)")
```
