Analysis and visualization of your **owned playlists**. Demonstrative only: sync is done via CLI or dashboard.

**Views:** Library overview, top artists, popularity distribution, release year timeline.

## Setup & load

```python
from pathlib import Path
from notebook_helpers import setup_project, get_data_dir, analyze_library, generate_library_statistics
from notebook_helpers import view_library_overview, view_top_artists, view_popularity_distribution, view_release_years

PROJECT_ROOT = setup_project(Path("../..").resolve())
DATA_DIR = get_data_dir(PROJECT_ROOT)

EXCLUDE_LIKED_SONGS = True
EXCLUDE_MONTHLY = False
INCLUDE_ONLY = []
EXCLUDE_PLAYLISTS = []

result = analyze_library(
    DATA_DIR,
    exclude_liked_songs=EXCLUDE_LIKED_SONGS,
    exclude_monthly=EXCLUDE_MONTHLY,
    include_only=INCLUDE_ONLY if INCLUDE_ONLY else None,
    exclude_names=EXCLUDE_PLAYLISTS if EXCLUDE_PLAYLISTS else None,
)
analyzer = result['analyzer']
```

## Library overview
